### exploración y limpieza del csv 'Books'

```python
df_books = pd.read_csv ('input_files/Books.csv')
df_books.head()
```

```python
df_books.columns
```

```python
df_books.columns = df_books.columns.str.lower().str.replace(' ', '_')
df_books
```

```python
df_books[df_books.duplicated()]
```

```python
df_books[df_books ['book'] == 'The Goal']
```

```python
df_books = df_books.drop_duplicates()
```

```python
df_books.shape
```

```python
df_books['approximate_sales'].unique()
```

```python
# Definir una función para limpiar los valores de la columna 'approximate_sales'
def clean_sales(value):
    # Utilizar expresión regular para encontrar el número en el formato deseado
    match = re.search(r'\d+', value)
    if match:
        return int(match.group())  # Convertir el valor encontrado a entero
    else:
        return None  # Devolver None si no se encuentra ningún número

# Aplicar la función a la columna 'approximate_sales' para limpiar los valores
df_books['approximate_sales'] = df_books['approximate_sales'].apply(clean_sales)
```

```python
df_books.to_csv ('output_files/books.csv')
```
